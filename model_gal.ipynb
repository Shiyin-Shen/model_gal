```python
from model_gal import dexp_SFH
from model_gal import exp_SFH
```

```python
time=np.arange(130)*0.1
```

```python
# SFH in unit MsunGyr^{-1}
SFH=np.zeros(130)
```

```python
for i in range(130):
    SFH[i]=dexp_SFH(time[i],tau=100.)
```

```python
plt.plot(time,SFH)
```

```python
mgal=model_gal.model_gal(dexp_SFH,velscale=70,FWHM_gal=3,vdisp=150,FeH=-2.,line_width=500.,lam_range=[3000,10000],tau=10)
```

```python
#spectra in unit of Lsun A^{-1}
air_wave=mgal.air_wave
stellar=mgal.stellar
gas_emission=mgal.gas_emission
spectra=mgal.spectra
plt.plot(air_wave,spectra)
plt.plot(air_wave,stellar)
plt.plot(air_wave,gas_emission)
plt.show()
```
